# Find the 3 richest people

Normally at a first sight, when i've looked at the 3 most valuable accounts value. I've got Ms Céline Lombard as the richest with these three different accounts

```python
top_richest = df.nlargest(3, 'g')
print(top_richest[['c', 'd', 'g']])
```

Then, i've grouped the final value (df['g']) by first and last name and finally i've got the three most richest people

```python
df['Total_Money'] = df.groupby(['c', 'd'])[['g']].transform('sum').sum(axis=1)

top_richest = df.drop_duplicates(subset=['c', 'd']).nlargest(3, 'Total_Money')

print(top_richest[['c', 'd', 'Total_Money']])
```
